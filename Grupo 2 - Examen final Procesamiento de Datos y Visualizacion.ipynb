**Presupuesto Nacional vs Presuesto del Ciudad de Buenos Aires: diferencias**

El Presupuesto Nacional es una ley anual que sanciona el Congreso, donde planifica los ingresos que recibir√° el Gobierno y c√≥mo se aplicar√°n para satisfacer las necesidades de la poblaci√≥n: seguridad social, educaci√≥n, salud, seguridad y justicia, entre otras. El Presupuesto financia el funcionamiento de todo el Estado Nacional; es decir de los Poderes Ejecutivo, Legislativo y Judicial. Tambi√©n contiene las transferencias que se realizan a los estados provinciales y municipales quienes a su vez confeccionan sus propios presupuestos.‚Äù

## üí∏**Variable Nominal vs Variable Real**: 
¬øC√≥mo comparar Presupuestos anuales en econom√≠as de alta inflaci√≥n? Para analizar y poder comparar los presupuestos, elegimos dos enfoques diferentes para:
- **IPC CABA**: el √çndice de Precios al Consumidor de CABA es una medida para los precios de los bienes y servicios adquiridos por un consumidor t√≠pico en la ciudad. Se calcula a en funci√≥n a una canasta (grupo de art√≠culos seg√∫n importancia/ponderaci√≥n de consumo) en relaci√≥n con un a√±o base (2011) para comparar su incremento con otros a√±os.  El IPC es una medida de nivel de precios (la tasa de inflaci√≥n es el cambio porcentual de los precios con respecto al a√±o o mes anterior).
Fuente: https://www.estadisticaciudad.gob.ar/eyc/?p=27386 
  
  Con el IPC promedio anual, deflactamos la serie, tomando como a√±o base 2020 y as√≠ comparar todo en una misma moneda actualizada por inflaci√≥n.

- **D√≥lar Tipo de Cambio**: si bien tomar el promedio de TC por a√±o nos trae una mejor conceptualizaci√≥n de la realidad dado que el peso argentino en valores nominales no podr√≠a compararse a lo largo del tiempo, no ser√≠a del todo correcto. #üí∞ **Tabla de valores de conversi√≥n de monedas**
Preparaci√≥n de una vista temporal con el c√°lculo de los √≠ndices de actualizaci√≥n a moneda 2021 por IPC y el valor del d√≥lar oficial al mes de sanci√≥n del presupuesto para cada uno de los cinco a√±os.

# üìë **Datasets para an√°lisis de presupuestos anuales**
En esta secci√≥n se desarrolla la selecci√≥n de los campos de cada per√≠odo de presupuesto combinando con los valores de las tablas adicionales para el an√°lisis.


```python
#presupuesto 2017
p2017 = spark.sql("SELECT 2017 as periodo, b.barrios, pre.fun_desc, replace(pre.sancion,'.','') sancion, p.total, replace(p.superficie,',','.') superficie, \
(select indice from monedas where periodo=2017) as indice, \
(select usd from monedas where periodo=2017) as usd \
from presu2017 pre \
left join barrios b on pre.geo=comunas \
left join poblacion p on pre.geo=comuna")
```

```python
#presupuesto 2018
p2018 = spark.sql("SELECT 2018 as periodo, b.barrios, pre.fun_desc, replace(pre.sancion,'.','') sancion, p.total, replace(p.superficie,',','.') superficie, \
(select indice from monedas where periodo=2018) as indice, \
(select usd from monedas where periodo=2018) as usd \
from presu2018 pre \
left join barrios b on pre.geo=comunas \
left join poblacion p on pre.geo=comuna")
```

```python
#presupuesto 2019
p2019 = spark.sql("SELECT 2019 as periodo, b.barrios, pre.fun_desc, replace(pre.sancion,',','') sancion, p.total, replace(p.superficie,',','.') superficie, \
(select indice from monedas where periodo=2019) as indice, \
(select usd from monedas where periodo=2019) as usd \
from presu2019 pre \
left join barrios b on pre.geo=comunas \
left join poblacion p on pre.geo=comuna")
```

```python
#presupuesto 2020
p2020 = spark.sql("SELECT 2020 as periodo, b.barrios, pre.fun_desc, replace(pre.sancion,'.','') sancion, p.total, replace(p.superficie,',','.') superficie, \
(select indice from monedas where periodo=2020) as indice, \
(select usd from monedas where periodo=2020) as usd \
from presu2020 pre \
left join barrios b on pre.geo=comunas \
left join poblacion p on pre.geo=comuna")
```

```python
#presupuesto 2021
p2021 = spark.sql("SELECT 2021 as periodo, b.barrios, pre.fun_desc, replace(pre.sancion,'.','') sancion, p.total, replace(p.superficie,',','.') superficie, \
(select indice from monedas where periodo=2021) as indice, \
(select usd from monedas where periodo=2021) as usd \
from presu2021 pre \
left join barrios b on pre.geo=comunas \
left join poblacion p on pre.geo=comuna")
```

```python
#concatenaci√≥n de los cinco presupuestos en un dataframe
presupuesto = p2017.union(p2018).union(p2019).union(p2020).union(p2021)
```

#üìù**C√°lculo de valores e indicadores**
El dataframe Spark con los presupuestos unificados se convierte a Pandas para procesarlo.

Se agregan columnas con el c√°lculo del presupuesto sancionado en moneda constante del a√±o 2021 y su valor en d√≥lares oficiales del mes de sanci√≥n.

Se calcula tambi√©n el presupuesto en moneda constante por habitante para cada comuna.




```python
#convierte dataframe spark a pandas 
# presupuesto: dataframe spark
# total: dataframe pandas
total = presupuesto.toPandas()
#ajuste de tipos de datos
total = total.astype({'sancion': 'float','superficie':'float'})
#c√°lculo de presupuesto sancionado en moneda 2021
total['constante'] = total['sancion']*total['indice']
#c√°lculo de presupuesto sancionado en d√≥lares
total['dolares'] = total['sancion']/total['usd']
#c√°lculo de presupuesto asignado por habitante, moneda 2021
total['por_habitante'] = total['constante']/total['total']
```
